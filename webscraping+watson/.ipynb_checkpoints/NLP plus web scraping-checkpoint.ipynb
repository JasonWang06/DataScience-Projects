```python
import json
from watson_developer_cloud import NaturalLanguageUnderstandingV1
import watson_developer_cloud.natural_language_understanding.features.v1 as Features
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
def getArticle(url):
    html = urlopen(url)
    soup = BeautifulSoup(html)
    article_text = ''
    article = soup.find("div", {"class":"article"}).findAll('p')
    for element in article:
        article_text += '\n' + ''.join(element.findAll(text = True))
    return article_text
```

```python
#scraping the article
article = getArticle("https://hbr.org/2002/05/why-business-models-matter")
```

```python
natural_language_understanding = NaturalLanguageUnderstandingV1(
    version="2017-02-27",
    username= "f5a94af4-8d30-4e17-ab0e-4dcf4c13334c",
    password="6F4aRO0rxkrd")
```

```python
#use watson to analyze the text
response = natural_language_understanding.analyze(
  url="https://hbr.org/2002/05/why-business-models-matter",
  features=[
    Features.Entities(),
    Features.Keywords(emotion = True),
    Features.Relations(),
    Features.Concepts()
    ]
)
```
